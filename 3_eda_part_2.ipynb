```python
import pandas as pd

df_prices_ww = pd.read_parquet('prices_ww_cleaned.parquet')
df_prices_ww
```

```python
df_prices_yf = pd.read_parquet('yf_prices.parquet')
df_prices_yf
```

```python
# акции с подозрительно низким количеством цен
value_counts = df_prices_yf['stock_id'].value_counts()
value_counts.tail(10)
```

```python
# удалим акции, по которым только 1 цена
value_counts[value_counts < 1].index
df_prices_yf = df_prices_yf[~df_prices_yf['stock_id'].isin(value_counts[value_counts < 5].index)]
df_prices_yf
```

```python
import pickle

with open('to_check.pkl', 'rb') as file:
    stock_ids_to_check = pickle.load(file)
stock_ids_to_check
```

```python
import nbformat
nbformat.__version__
```

```python
# ручная валидация цен

import plotly.graph_objects as go

for stock_id in stock_ids_to_check:
    price_ww = df_prices_ww[df_prices_ww['stock_id'] == stock_id]
    price_yf = df_prices_yf[df_prices_yf['stock_id'] == stock_id]

    if not price_yf.empty:
        ticker = price_yf.iloc[0]['yf_ticker']
    else:
        ticker = 'not found on YF'
    stock_name = price_ww.iloc[0]['name']
    stock_permalink = price_ww.iloc[0]['stock_permalink']

    # оставить только показательные графики - отобраны вручную без этого фильтра
    if ticker not in 'MTEM VTAK APTO LPTX GENE KRRO QTTB DRMA CRBP CKPT BNTC AWH QNRX TOVX'.split() and stock_permalink != 'mgta':
        continue

    trace_ww = go.Scatter(x=price_ww['date'], y=price_ww['closing_price'], mode='lines', name='ww', line=dict(width=6, color='cyan'))
    trace_yf = go.Scatter(x=price_yf['date'], y=price_yf['closing_price'], mode='lines', name='yf', line=dict(width=3, color='red'))

    layout = go.Layout(
        title=f'{ticker} ({stock_name}, {stock_permalink})',
        plot_bgcolor='#1e2130' if not price_yf.empty else '#3b0000',
        paper_bgcolor='#1e2130' if not price_yf.empty else '#3b0000',
        xaxis=dict(title='Date', color='white'),
        yaxis=dict(range=[0, None], title='Closing Price', color='white'),
        font=dict(color='white')
    )

    fig = go.Figure(data=[trace_ww, trace_yf], layout=layout)
    fig.show()
```

```python
# ручной фикс цен, кривых в обоих источниках

from datetime import datetime

for stock_id in stock_ids_to_check:
    price_ww = df_prices_ww[df_prices_ww['stock_id'] == stock_id]
    price_yf = df_prices_yf[df_prices_yf['stock_id'] == stock_id]

    if not price_yf.empty:
        ticker = price_yf.iloc[0]['yf_ticker']
    else:
        ticker = 'not found on YF'
    stock_id = price_ww.iloc[0]['stock_id']
    stock_name = price_ww.iloc[0]['name']
    stock_permalink = price_ww.iloc[0]['stock_permalink']

    traces = []

    if ticker == 'APRE':
        print('обратный сплит 13.02.2023 (1 for 20)')
        traces.append(go.Scatter(x=price_yf['date'], y=price_yf['closing_price'], mode='lines', name='цена yf до ручного фикса', line=dict(width=3, color='red')))
        df_prices_yf.loc[(df_prices_yf['stock_id'] == stock_id) & (df_prices_yf['date'] == datetime(year=2023, month=2, day=13)), 'closing_price'] *= 20
        price_yf.loc[(price_yf['stock_id'] == stock_id) & (price_yf['date'] == datetime(year=2023, month=2, day=13)), 'closing_price'] *= 20
    elif ticker == 'TECX':
        print('обратный сплит 21.06.2024 (1 for 12)')
        traces.append(go.Scatter(x=price_yf['date'], y=price_yf['closing_price'], mode='lines', name='цена yf до ручного фикса', line=dict(width=3, color='red')))
        df_prices_yf.loc[(df_prices_yf['stock_id'] == stock_id) & (df_prices_yf['date'] == datetime(year=2024, month=6, day=21)), 'closing_price'] *= 12
        price_yf.loc[(price_yf['stock_id'] == stock_id) & (price_yf['date'] == datetime(year=2024, month=6, day=21)), 'closing_price'] *= 12
    elif ticker == 'MBRX':
        print('обратный сплит 22.03.2024 (1 for 15)')
        traces.append(go.Scatter(x=price_yf['date'], y=price_yf['closing_price'], mode='lines', name='цена yf до ручного фикса', line=dict(width=3, color='red')))
        df_prices_yf.loc[(df_prices_yf['stock_id'] == stock_id) & (df_prices_yf['date'] <= datetime(year=2024, month=3, day=21)), 'closing_price'] *= 15
        price_yf.loc[(price_yf['stock_id'] == stock_id) & (price_yf['date'] <= datetime(year=2024, month=3, day=21)), 'closing_price'] *= 15
    elif stock_id == 206664:
        print('обратный сплит 09.11.2023 (1 for 16)')
        traces.append(go.Scatter(x=price_ww['date'], y=price_ww['closing_price'], mode='lines', name='цена ww до ручного фикса', line=dict(width=3, color='red')))
        df_prices_ww.loc[(df_prices_ww['stock_id'] == stock_id) & (df_prices_ww['date'] <= datetime(year=2023, month=9, day=11)), 'closing_price'] *= 16
        price_ww.loc[(price_ww['stock_id'] == stock_id) & (price_ww['date'] <= datetime(year=2023, month=9, day=11)), 'closing_price'] *= 16
    elif stock_id == 241886:
        print('обратный сплит 21.11.2024 (1 for 4)')
        traces.append(go.Scatter(x=price_yf['date'], y=price_yf['closing_price'], mode='lines', name='цена yf до ручного фикса', line=dict(width=3, color='red')))
        df_prices_yf.loc[(df_prices_yf['stock_id'] == stock_id) & (df_prices_yf['date'] <= datetime(year=2024, month=11, day=20)), 'closing_price'] *= 4
        price_yf.loc[(price_yf['stock_id'] == stock_id) & (price_yf['date'] <= datetime(year=2024, month=11, day=20)), 'closing_price'] *= 4
    else:
        continue

    traces.append(go.Scatter(x=price_ww['date'], y=price_ww['closing_price'], mode='lines', name='ww', line=dict(width=6, color='cyan')))
    traces.append(go.Scatter(x=price_yf['date'], y=price_yf['closing_price'], mode='lines', name='yf', line=dict(width=3, color='chartreuse')))

    layout = go.Layout(
        title=f'{ticker} ({stock_name}, {stock_permalink}, {stock_id})',
        plot_bgcolor='#1e2130' if not price_yf.empty else '#3b0000',
        paper_bgcolor='#1e2130' if not price_yf.empty else '#3b0000',
        xaxis=dict(title='Date', color='white'),
        yaxis=dict(range=[0, None], title='Closing Price', color='white'),
        font=dict(color='white')
    )

    fig = go.Figure(data=traces, layout=layout)
    fig.show()

```

```python
# теперь все цены, собранные из yahoo finance, ввлидированы, перезапишем ими цены whalewisdom
# для акций, не найденных на yahoo finance, валидированы все цены whalewisdom

# мерджим датафреймы следующим образом:
# создаем новый df из df_prices_ww содержащий только stock_ids которые есть в df_prices_yf
# удаляем эти строки из df_prices_ww
# делаем inner join между новым df и df_prices_yf
# добавляем результат к очищенному df_prices_ww

new_df_ww = df_prices_ww[df_prices_ww['stock_id'].isin(df_prices_yf['stock_id'])]
df_prices_ww_cleaned = df_prices_ww[~df_prices_ww['stock_id'].isin(df_prices_yf['stock_id'])]
inner_join_df = new_df_ww.merge(df_prices_yf, on=['date', 'stock_id'], suffixes=('', '_yf'))
inner_join_df['closing_price'] = inner_join_df['closing_price_yf']
inner_join_df.drop(columns='closing_price_yf', inplace=True)
final_df = pd.concat([df_prices_ww_cleaned, inner_join_df], ignore_index=True)

# проверяем, что все отработало корректно, мы ничего не потеряли и данных меньше не стало
print(df_prices_ww.shape)
print(final_df.shape)

# удалим high, open, low, чтобы не путаться, так как их не валидировали
final_df.drop(columns=['high', 'open', 'low'], inplace=True)

final_df.to_parquet('merged_prices.parquet')
```
